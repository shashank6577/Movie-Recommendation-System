```python
import numpy as np
```

```python
import pandas as pd
```

```python
movie=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie =movie.merge(credit,on='title')
```

```python
movie=movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie
```

```python
movie.dropna(inplace=True)
```

```python
movie.duplicated().sum()
```

```python
movie.iloc[0,3]
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convert2(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3 :
            count=count+1
            L.append(i['name'])
        else:
            break
    return L
```

```python
movie["cast"]=movie["cast"].apply(convert2)
```

```python
movie['cast']
```

```python
def director(obj):
    r=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            r.append(i['name'])
            break
    return r
```

```python
movie['crew']=movie['crew'].apply(director)
```

```python
movie
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie['tags']=movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
new_df=movie[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
pip install scikit-learn
```

```python
 from  sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
new_df['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movies):
    movie_idx=new_df[new_df['title']==movies].index[0]
    distances=similarity[movie_idx]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Superman')
```
